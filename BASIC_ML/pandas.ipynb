menghapus missing value pada dataframe (Nan)

```python
def make_missing_dataframe(nrows=30, ncols=4):
    data = np.random.randn(nrows, ncols)
    df = pd.DataFrame(data, columns=[cols[i] for i in range(ncols)])

    # Randomly assign NaN values
    for col in df.columns:
        df.loc[df.sample(frac=0.1).index, col] = np.nan

    return df

df = make_missing_dataframe()

# persentase missing value
df.isna().mean()
```

```python
df_backup = df.copy(deep=True)
print(df)

df = df.dropna(axis='columns')
df
```

```python
df = df_backup.copy(deep=True)
print(df)
df = df.dropna(axis='rows')
df
```

```python
print(df)
df = df_backup.copy(deep=True)
print(df)
treshold = len(df) * 0.01
df = df.dropna(thresh=treshold, axis='columns')
df
```
