Pemilihan baris (rows selection) pada Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns=cols)
df
```

Selector dengan operator logika "|"

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

Selection menggunakan fungsi isin()

```python
df[df["A"].isin([1, 3])]
```

Operator Negasi ~

```python
df[~df['A'].isin([1, 3])]
```

Konversi  tipe data String ke numerik pada kolom dataframe

```python
data = {"col1": ['1', '2', '3', 'teks'], "col2": ['4', '5', '6', '7']}

df = pd.DataFrame(data)
df
```

Konversi tipe data numerik dengan fungsi to_numeric()

```python
df.apply(pd.to_numeric, errors='coerce')
```

Pemilihan kolom (columns selection) pada Data Frame berdasarkan tipe data

```python


n_rows = 5
n_cols = 2

cols = ['bil_pecahan', 'bil_bulat']
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df['bil_pecahan'] = df['bil_pecahan'].astype('float')

dt = pd.date_range('today', periods=(n_rows), freq='h')
df.index = pd.to_datetime(dt, format='%d%b%Y:%H:%M:%S.%f')

df['teks'] = list('ABCDE')
df
```
